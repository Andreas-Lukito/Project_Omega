```python
from faker import Faker
import numpy as np
import random
import pandas as pd
```

```python
name_list = []
rating_list = []
category_list = []
variant_list = []
variant_price_list = []

category = ['Americano', 'Cappuchino', 'Latte', 'Frappuchino', 'Macchiato']
variant = ['vanilla', 'hazelnut', 'cinammon', 'butterscotch', 'caramel', 'salted caramel']
fake = Faker()

for i in range(0, 100):
    name_list.append(fake.name())
    rating_list.append(random.randrange(0,5))
    category_list.append(random.choice(category))
    variant_list.append(random.choice(variant))
    variant_price_list.append(random.randrange(10000, 60000))
    
```

```python
data = {
    'name' : name_list,
    'rating' : rating_list,
    'category': category_list,
    'variant' : variant_list,
    'variant_price' : variant_price_list
}
             
DataFrame = pd.DataFrame(data)
DataFrame.to_csv('data_for_database.csv')
```

```python
#generate time data
synthetic_data = pd.read_csv('synthetic_data.csv')
```

```python
import datetime

start_date = datetime.datetime(2024, 12, 10, 7, 0 , 0)
dates = []

current = start_date

for i in range(0,len(synthetic_data)): 
    dates.append(current)
    if(current.hour == 22):
        current = datetime.datetime(current.year, current.month, current.day, 7, 0, 0) + datetime.timedelta(days = 1)
    else:
        current += datetime.timedelta(hours = 1)
    print(current)
        
```

```python
synthetic_data['DateTime'] = dates
```

```python
# reorder data
synthetic_data = synthetic_data[['DateTime'] + [col for col in synthetic_data.columns if col != 'DateTime']]
```
